```python
import pandas as pd
import requests

# 1. Download UN file (manual download if direct URL fails)
url = "https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2022_TotalPopulationBySex.csv"
fname = "WPP2022_TotalPopulationBySex.csv"

# If needed, download
# r = requests.get(url)
# with open(fname, "wb") as f:
#     f.write(r.content)

# 2. Load dataset
df = pd.read_csv(fname)

# 3. Define ASEAN
asean_countries = [
    "Brunei Darussalam",
    "Cambodia",
    "Indonesia",
    "Lao People's Democratic Republic",
    "Malaysia",
    "Myanmar",
    "Philippines",
    "Singapore",
    "Thailand",
    "Viet Nam"
]

# 4. Filter ASEAN + Medium variant
asean_df = df[
    (df["Location"].isin(asean_countries)) &
    (df["Variant"] == "Medium") &
    (df["Time"] >= 1950) & (df["Time"] <= 2100)
]

# 5. Convert thousands → millions
asean_df["PopulationMillions"] = asean_df["PopTotal"] / 1000

# 6. Pivot to wide format
pivot_df = asean_df.pivot(index="Time", columns="Location", values="PopulationMillions").reset_index()
pivot_df.rename(columns={"Time": "Year"}, inplace=True)

# 7. ASEAN total
pivot_df["ASEAN Total"] = pivot_df[asean_countries].sum(axis=1)

# 8. Export
pivot_df.to_csv("asean_population_1950_2100.csv", index=False)
print("✅ Saved asean_population_1950_2100.csv")
```
